```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create scatter plot
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

- Scatter plots are useful for visualizing the relationship between two continuous variables.
- `sns.scatterplot()` is used to create a scatter plot in Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create line plot
sns.lineplot(x=x, y=y)
plt.title('Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

- Line plots are used to visualize the trend of a continuous variable over another continuous variable.
- `sns.lineplot()` creates a line plot in Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
data = [1, 1, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 6]

# Create histogram
sns.histplot(data, bins=5)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

- Histograms are used to visualize the distribution of a single numerical variable.
- `sns.histplot()` creates a histogram in Seaborn. You can specify the number of bins using the bins parameter.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
data = [1, 1, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 6]

# Create box plot
sns.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Value')
plt.show()
```

- Box plots are useful for visualizing the distribution of numerical data through their quartiles.
- `sns.boxplot()` creates a box plot in Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Create bar plot
sns.barplot(x=categories, y=values)
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

- Bar plots are used to visualize the distribution of a categorical variable.
- `sns.barplot()` creates a bar plot in Seaborn.

### Subplot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot bar plot
axs[0].bar(x, y)
axs[0].set_title('Bar Plot')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Plot scatter plot
axs[1].scatter(x, y)
axs[1].set_title('Scatter Plot')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

plt.tight_layout()
plt.show()

```

- difference between axes and figure? 
- axes is used to create a subplot, while figure is used to create a figure.
