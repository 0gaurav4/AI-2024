```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample data 
x = [1, 2, 3, 4, 5]
y = [2, 5, 7, 1, 8]

# Create a line plot
sns.lineplot(x=x, y=y)

# Displaying the plot
plt.show()
```

```python
sns.lineplot(x=x, y=y)
# Customize the plot (optional)
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Line Plot Example")
plt.grid(True)  # Add gridlines

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting the data using Seaborn
sns.scatterplot(x=x, y=y)

# Displaying the plot
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [10, 20, 15, 25]

# Creating a bar plot
sns.barplot(x=x, y=y)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Mean Value')
plt.title('Bar Plot')

# Displaying the plot
plt.show()
```

```python
data = ['A', 'B', 'A', 'A', 'C', 'B', 'B', 'C', 'C', 'C']

# Creating a count plot
sns.countplot(x=data)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count Plot')

# Displaying the plot
plt.show()
```

# Some basic plots using datasets

Scatter Plot:
Use the 'iris' dataset from Seaborn's built-in datasets to create a scatter plot.

```python
# Load the 'iris' dataset
iris = sns.load_dataset('iris')

# Create a scatter plot
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width')

# Adding labels and title
# plt.xlabel('Sepal Length')
# plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')

# Displaying the plot
plt.show()
```

```python
# Load the 'iris' dataset
iris = sns.load_dataset('iris')

# Create a scatter plot
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')

# Adding labels and title
# plt.xlabel('Sepal Length')
# plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')

# Displaying the plot
plt.show()
```

```python
# Generate sample data
data = np.random.randn(1000)

# Create a histogram
sns.histplot(data)

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

# Displaying the plot
plt.show()
```

```python
# Generate sample data
data = np.random.randn(1000)

# Create a histogram
sns.histplot(data, kde=True)

# Adding labels and title

plt.title('Histogram')

# Displaying the plot
plt.show()
```

Bar Plot:We'll use the 'tips' dataset from Seaborn's built-in datasets to create a bar plot.

```python
# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Create a bar plot
sns.barplot(data=tips, x='day', y='total_bill')

# Adding labels and title
plt.title('Bar Plot of Total Bill by Day and Gender')

# Displaying the plot
plt.show()
```

```python
# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Create a bar plot
sns.barplot(data=tips, x='day', y='total_bill', hue='sex', ci=None)

# Adding labels and title
plt.title('Bar Plot of Total Bill by Day and Gender')

# Displaying the plot
plt.show()
```

Box Plot:
We'll use the 'tips' dataset from Seaborn's built-in datasets to create a box plot.

```python
# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Create a box plot
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')

# Adding labels and title
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Box Plot of Total Bill by Day and Gender')

# Displaying the plot
plt.show()
```
