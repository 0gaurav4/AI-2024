# 1. Relational Plots

```python
# Load sample dataset
tips = sns.load_dataset("tips")

# Create a relational plot (scatter plot)
sns.relplot(data=tips, x="total_bill", y="tip",kind="scatter")

# Customizing labels and title
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Relationship between Total Bill and Tip')

# Displaying the plot
plt.show()
```

```python
# Create a relational plot (scatter plot) with customization
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex")

# Customizing labels and title
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Relationship between Total Bill and Tip')

# Displaying the plot
plt.show()
```

```python
# Create a relational plot (scatter plot) with customization
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", style="time")

# Customizing labels and title
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Relationship between Total Bill and Tip')

# Displaying the plot
plt.show()
```

```python
# Load sample dataset
tips = sns.load_dataset("tips")

# Create a relational plot (scatter plot) with customization
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", size="size")
# Customizing labels and title
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Relationship between Total Bill and Tip')

# Displaying the plot
plt.show()
```

```python
# Load sample dataset
tips = sns.load_dataset("tips")

# Create a relational plot (scatter plot) with customization
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", size="size", palette="deep")

# Customizing labels and title
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Relationship between Total Bill and Tip')

# Displaying the plot
plt.show()
```

# 2. Categorical plots

```python
# Create a strip plot
sns.catplot(data=tips, x="day", y="total_bill", kind="strip")

# Display the plot
plt.show()
```

```python
# Create a swarm plot
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")

# Display the plot
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box")

# Display the plot
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="violin")

# Display the plot
plt.show()
```

```python
sns.catplot(data=tips, x="day", hue="sex", kind="count")

# Display the plot
plt.show()
```

# 3. Distribution Plots

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist',bins=4)
```

```python
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
```

```python
# element -> step
sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```
