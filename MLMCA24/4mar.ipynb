```python
#we have to find out Value count of male & female survivor? 

''' df['Survived'] == 1 is used as a condition to filter the DataFrame df.
This condition checks if the values in the 'Survived' column are equal to 1.

'Survived' column contains binary values
indicating whether a person survived (1) or not (0). By using == 1,
the code selects only the rows where the 'Survived' column has a value of 1, indicating the survivors.'''

survivor_counts = df[df['Survived'] == 1]['Sex'].value_counts()
print(survivor_counts)
```
