```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic_data = pd.read_csv('../data/titanic.csv')
titanic_data.head()
```

```python
#to rename column names
titanic_data.rename(columns={' Age ':'Age'}, inplace=True)
titanic_data.rename(columns={' SibSp':'SibSp'}, inplace=True)
titanic_data.rename(columns={' Parch':'Parch'}, inplace=True)
titanic_data.rename(columns={' Embarked':'Embarked'}, inplace=True)
titanic_data.rename(columns={' Pclass':'Pclass'}, inplace=True)
titanic_data.rename(columns={' Survived':'Survived'}, inplace=True)
titanic_data.rename(columns={' Fare    ':'Fare'}, inplace=True)
titanic_data.rename(columns={' Sex   ':'Sex'}, inplace=True)
titanic_data.rename(columns={' Ticket            ':'Ticket'}, inplace=True)
titanic_data.rename(columns={' Cabin          ':'Cabin'}, inplace=True)
titanic_data.rename(columns={' Name                                                                                ':'Name'}, inplace=True)
```

# Bar Plot:
Show the count of passengers who survived and those who didn't.

```python
# Count survived vs not survived
survived_count = titanic_data['Survived'].value_counts()

# Plotting
plt.bar(survived_count.index.astype(str), survived_count.values)
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.title('Survival Count')
plt.show()
```

# Histogram:
Show the distribution of passenger ages.

```python
# Drop missing values in 'age' column
age_data = titanic_data['Age'].dropna()

# Plotting
plt.hist(age_data, bins=20, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Passenger Age Distribution')
plt.show()
```

# Pie Chart:
Show the distribution of passengers by class.

```python
# Count passengers by class
class_counts = titanic_data['Pclass'].value_counts()

# Plotting
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Passenger Class Distribution')
plt.show()
```

# Scatter Plot:
Show the relationship between fare and age of passengers.

```python
# Plotting
plt.scatter(titanic_data['Age'], titanic_data['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs Age')
plt.show()
```

# Bar Plot with Grouping:
Show the count of passengers who survived or not survived based on gender.

```python
titanic_data.groupby(['Sex', 'Survived']).size().unstack()
```

```python
# Grouping data by 'sex' and 'survived' columns
survived_by_sex = titanic_data.groupby(['Sex', 'Survived']).size().unstack()

# Plotting
survived_by_sex.plot(kind='bar', stacked=True)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survival Count by Gender')
plt.legend(['Not Survived', 'Survived'], loc='upper left')
plt.show()
```
